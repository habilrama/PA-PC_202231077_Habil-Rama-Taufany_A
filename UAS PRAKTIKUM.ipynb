## Median Filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('Pawai SMP.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
median_filtered = cv2.medianBlur(image_gray, 5)
```

```python
#Menampilkan gambar asli dan hasil median filtering
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Citra asli')
plt.imshow(image_rgb)

plt.subplot(1, 2, 2)
plt.title('After Filter Median')
plt.imshow(median_filtered, cmap='gray')

plt.show()
```

## Mean Filtering

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Fungsi untuk mean
def mean_filtering(image, kernel_size):
    # Padding gambar
    pad_size = kernel_size // 2
    padded_image = np.pad(image, pad_size, mode='constant', constant_values=0)

    # Inisialisasi hasil filter
    mean_filtered = np.zeros_like(image)

    # Iterasi setiap piksel
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):

            # Ekstrasi neighborhodd
            neighborhood = padded_image[i:i+kernel_size, j:j+kernel_size]
            # Rata rata nilai piksel
            mean_filtered [i, j] = np.mean(neighborhood)
            
    return mean_filtered
```

```python
image = cv2.imread('Pawai SMP.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
mean_filtered = mean_filtering(image, 5)
```

```python
# Menampilkan gambar asli dan mean filtering
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Mean Filtered Image')
plt.imshow(mean_filtered, cmap='gray')

plt.show()
```
